```python
# Count the occurrences of each unique value in the 'Category' column
value_counts = df['Class'].value_counts()

# Plot the counts
plt.figure(figsize=(10, 6))
value_counts.plot(kind='bar', color='skyblue')

# Add titles and labels
plt.title('Count of Each Unique Value in Category Column')
plt.xlabel('Fraud')
plt.ylabel('Count')

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18,4))

amount_val = df['Amount'].values
time_val = df['Time'].values

sns.distplot(amount_val, ax=ax[0], color='r')
ax[0].set_title('Distribution of Transaction Amount', fontsize=14)
ax[0].set_xlim([min(amount_val), max(amount_val)])

sns.distplot(time_val, ax=ax[1], color='b')
ax[1].set_title('Distribution of Transaction Time', fontsize=14)
ax[1].set_xlim([min(time_val), max(time_val)])



plt.show()
```

```python
# split the data into testing and training data
# here y represents target variable (for stratification) and X represents feature variable
y = df["Class"]
X = df.drop("Class", axis="columns")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=23)

# Normalize amount - need to only fit on training data, therefore X_train. Now all columns are minmax scaled
min_max_scaler = MinMaxScaler()
X_train_scaled = min_max_scaler.fit_transform(X_train)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
```

```python
# Visualize correlation between features by creating a seaborn heatmap

correlation_mtx = X_train_scaled.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_mtx, annot=True, fmt=".2f", cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.show()
```
